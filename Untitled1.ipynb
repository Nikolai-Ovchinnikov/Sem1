```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
x = [1,2,3,4,5,6]
y = [20,30,40,50,60,70]
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,color='green',marker='^')
plt.title('график')
ax = plt.gca()
ax.set_xlim(1,8)
ax.set_ylim(20,70)
ax.set_xticks([1,2,3,4,5,6,6.5])
ax.set_yticks([20,30,40,50,60,70,80])
plt.grid(True)
plt.show()
```

```python
spisok=[4,3,4,5,5]
dni = ['понедельник',"вторник","среда","четверг","пятница"]
plt.pie(spisok, labels=dni, autopct='%1.f%%')
plt.title("распределение пар по дням недели")
plt.show()
```

```python
x = ['понедельник',"вторник","среда","четверг","пятница"]
y = [4, 3, 4, 5, 5]
plt.bar(x, y, label='распределение пар по неделе', alpha=0.6 )
plt.plot(x, y, color='red', marker='>', markersize=4)
plt.xlabel('дни недели')
plt.ylabel('количество пар')
plt.title('распределение пар по неделе')
plt.legend(loc='lower left')
plt.show()
```
